```python
from skimage.draw import polygon
from skimage.draw import polygon_perimeter
import numpy as np
import matplotlib.pyplot as plt
def hexgrid(arcseconds, x, y):
    a = arcseconds 
    radi = (a/.03)/ 2
    img = np.zeros((25, 25), dtype=np.uint16)
    height = radi * np.cos(30 * np.pi / 180)
    width = radi * np.sin(30 * np.pi/180)
    c = np.array([ x - radi, x - width, x + width, x + radi, x + width, x - width, x - radi])
    r = np.array([ y , y + height, y + height, y, y - height, y - height, y])
    rr, cc = polygon(r, c)
#    yy, xx = polygon_perimeter(r,c)
#    rr = np.append([rr],[yy])
#    cc = np.append([cc],[xx])
    img[rr, cc] = 1
    print(c[0])
    area = (.5 * height * radi) * 6
    print(area)
    print(np.sum(img))
    return(img)

hex1 = hexgrid(.2, 12, 12)
hex2 = hexgrid()
#Notes
# arcseconds < .1 area of 5 isn't a hexagon
# .1<= arc <= .13 3x3 and 1 pixel on the left and right 
```
